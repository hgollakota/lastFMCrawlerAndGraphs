```python
#REQUIRES the LATEST networkx 2.2. Will be using a method only available in 2.2
import networkx as nx
#check networkx version by uncommenting following command:
#nx.__version__
import matplotlib.pyplot as plt
%matplotlib inline 
#%autoreload     # This is Jupyter notebook that will reload changed files
from crawler import Crawler
from crawler_state import CrawlerState
from crawler_abst_api import CrawlerAbstractAPI
from crawler_my_api import MyAPI
import pandas as pd
import seaborn as sns
```

```python
api = MyAPI()
craw = Crawler(5, api)
craw.initialize()
craw.set_autosave(10, "C:/Users/Harry/Desktop/11-12_projectcode/60s_5d_2000_41init/")
```

```python
craw.crawl_k(2000)
```

```python
#Makes craw._net actually bipartite
#removes self loops
selfloops = list(nx.selfloop_edges(craw._net))
craw._net.remove_edges_from(selfloops)
#identifies top and bottom nodes
top_nodes = {n for n, d in craw._net.nodes(data=True) if d['bipartite']==0}
bottom_nodes = {n for n, d in craw._net.nodes(data=True) if d['bipartite']==1}
#identifies edges that exist between either the top nodes or bottom nodes
temptop = nx.subgraph(craw._net, top_nodes)
tempbottom = nx.subgraph(craw._net, bottom_nodes)
topedges = list(temptop.edges())
bottomedges = list(tempbottom.edges())
#removes edges between nodes of type A or top and edges between nodes of type B or bottom
craw._net.remove_edges_from(topedges)
craw._net.remove_edges_from(bottomedges)
#Tests to see if the network is bipartite
nx.is_bipartite(craw._net)
```

```python
#add shortest path attribute
shortestPath60s = nx.single_source_shortest_path_length(craw._net, '100')
nx.set_node_attributes(craw._net, shortestPath60s, 'BipartiteShortestPathTo60sNode')
```

```python
#Add bipartite clustering attribute
bip_clust_coef = nx.algorithms.bipartite.cluster.clustering(craw._net)
nx.set_node_attributes(craw._net, bip_clust_coef, 'BipartiteClusteringCoefficient')
```

```python
#Add bipartite closeness centrality. Might take several minutes to calculate
bip_cent_closeness = nx.algorithms.bipartite.centrality.closeness_centrality(craw._net, top_nodes)
nx.set_node_attributes(craw._net, bip_cent_closeness, 'BipartiteClosenessCentrality')
```

```python
#Add bipartite degree centrality. Might take several minutes to calculate
bip_cent_degree = nx.algorithms.bipartite.centrality.degree_centrality(craw._net, top_nodes)
nx.set_node_attributes(craw._net, bip_cent_degree, 'BipartiteDegreeCentrality')
```

```python
#Add bipartite betweenness centrality. Might take several minutes to calculate
bip_cent_bet = nx.algorithms.bipartite.centrality.betweenness_centrality(craw._net, top_nodes)
nx.set_node_attributes(craw._net, bip_cent_bet, 'BipartiteBetweennessCentrality')
```

```python
#Calculate robins alexander clustering
#measures an overall level of clustering in the network
#for more information on the topic I looked into https://toreopsahl.com/tnet/two-mode-networks/clustering/
nx.algorithms.bipartite.cluster.robins_alexander_clustering(G=craw._net)
```

```python
#Calculates the bipartite density
nx.algorithms.bipartite.density(craw._net, top_nodes)
```

```python
#Calculates the diameter of the full bipartite graph
nx.algorithms.distance_measures.diameter(craw._net)
```

```python
#Calculate greedy modularity communities of the entire bipartite graph
#new feature of networkx 2.2
#based on a 2004 paper by Clauset, Newman and Moore http://ece-research.unm.edu/ifis/papers/community-moore.pdf
#Calculates the greedy modularity communities for the full graph
from networkx.algorithms import community
from networkx.algorithms.community import greedy_modularity_communities

full60s_communities = list(greedy_modularity_communities(craw._net))
len(full60s_communities)
```

```python
#Adds the community attribute to every node
bipartiteCommunityDict = {}
communityNum = 1
for community in full60s_communities:
    for nodename in community:
        bipartiteCommunityDict[nodename] = communityNum
    communityNum = communityNum + 1
nx.set_node_attributes(craw._net, bipartiteCommunityDict, 'BipartiteCommunity')
```

```python
#Creates two weighted projected graphs
artist60s_nodes = [node for node, node_data in craw._net.nodes(data=True) if node_data['bipartite']==False]
tags60s_nodes = [node for node, node_data in craw._net.nodes(data=True) if node_data['bipartite']==True]
artists60s = nx.bipartite.weighted_projected_graph(craw._net, artist60s_nodes)
tags60s = nx.bipartite.weighted_projected_graph(craw._net, tags60s_nodes)
```

```python
#Calculates average clustering (approximation)
from networkx.algorithms import approximation

aclust = nx.algorithms.approximation.clustering_coefficient.average_clustering(artists60s)
tclust = nx.algorithms.approximation.clustering_coefficient.average_clustering(tags60s)
print("Average clustering for Artist-to-Artist 60s :")
print(aclust)
print("Average clustering for Tag-to-Tag 60s :")
print(tclust)
```

```python
#Calculate density for both projected graphs
adens = nx.density(artists60s)
tdens = nx.density(tags60s)
print("Artist-to-Artist 60s Density: ")
print(adens)
print("Tag-to-Tag 60s Density: ")
print(tdens)
```

```python
#Calculate diameter of both projected graphs
adiam = nx.algorithms.distance_measures.diameter(artists60s)
tdiam = nx.algorithms.distance_measures.diameter(tags60s)

print("Artist-to-Artist 60s diameter:")
print(adiam)
print("Tag-to-Tag 60s diameter: ")
print(tdiam)
```

```python
artist60s_communities = list(greedy_modularity_communities(artists60s))
tag60s_communities = list(greedy_modularity_communities(tags60s))

print("Number communities in Artist-to-Artist projection: ")
print(len(artist60s_communities))
print("Number of communities in Tag-to-Tag projection: ")
print(len(tag60s_communities))
```

```python
#Adds the community attribute to every node
artCommunityDict = {}
artcommunityNum = 1
for community in artist60s_communities:
    for nodename in community:
        artCommunityDict[nodename] = artcommunityNum
    artcommunityNum = artcommunityNum + 1
nx.set_node_attributes(craw._net, artCommunityDict, 'ProjectedCommunity')
nx.set_node_attributes(artists60s, artCommunityDict, 'ProjectedCommunity')

tagCommunityDict = {}
tagcommunityNum = 1
for community in tag60s_communities:
    for nodename in community:
        tagCommunityDict[nodename] = tagcommunityNum
    tagcommunityNum = tagcommunityNum + 1
nx.set_node_attributes(craw._net, tagCommunityDict, 'ProjectedCommunity')
nx.set_node_attributes(tags60s, tagCommunityDict, 'ProjectedCommunity')
```

```python
#Writes the full bipartite graph to gexf and graphml
nx.write_gexf(craw._net, "60s_5depth_2000.gexf")
nx.write_graphml(craw._net, "60s_5depth_2000.graphml")
```

```python
#Writes the individual projected graphs to gexf and graphml
nx.write_gexf(artists60s, "artists60s_5d_2000.gexf")
nx.write_graphml(artists60s, "artists60s_5d_2000.graphml")
nx.write_gexf(tags60s, "tags60s_5d_2000.gexf")
nx.write_graphml(tags60s, "tags60s_5d_2000.graphml")
```

```python
#Sizes of each network
print("60s bipartite size: ", len(craw._net))
print("60s A2AP size: ", len(artists60s))
print("60s T2TP size: ", len(tags60s))
print(len(tags60s)+len(artists60s))
```
