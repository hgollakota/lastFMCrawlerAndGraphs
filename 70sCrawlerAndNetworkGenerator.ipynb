```python
#REQUIRES the LATEST networkx 2.2. Will be using a method only available in 2.2
import networkx as nx
#check networkx version by uncommenting following command:
print(nx.__version__)
import matplotlib.pyplot as plt
%matplotlib inline 
#%autoreload     # This is Jupyter notebook that will reload changed files
from crawler import Crawler
from crawler_state import CrawlerState
from crawler_abst_api import CrawlerAbstractAPI
from crawler_my_api import MyAPI
import pandas as pd
import seaborn as sns
```

```python
api2 = MyAPI()
craw2 = Crawler(5, api2)
craw2.initialize()
craw2.set_autosave(10, "C:/Users/Harry/Desktop/11-12_projectcode/70s_5d_2000_41init/")
```

```python
craw2.crawl_k(2000)
```

```python
#makes a copy of craw2._net just in case
justcrawled_craw2 = craw2._net.copy()
```

```python
#Makes craw2._net actually bipartite
#removes self loops
selfloops = list(nx.selfloop_edges(craw2._net))
craw2._net.remove_edges_from(selfloops)
#identifies top and bottom nodes
top_nodes = {n for n, d in craw2._net.nodes(data=True) if d['bipartite']==0}
bottom_nodes = {n for n, d in craw2._net.nodes(data=True) if d['bipartite']==1}
#identifies edges that exist between either the top nodes or bottom nodes
temptop = nx.subgraph(craw2._net, top_nodes)
tempbottom = nx.subgraph(craw2._net, bottom_nodes)
topedges = list(temptop.edges())
bottomedges = list(tempbottom.edges())
#removes edges between nodes of type A or top and edges between nodes of type B or bottom
craw2._net.remove_edges_from(topedges)
craw2._net.remove_edges_from(bottomedges)
#Tests to see if the network is bipartite
nx.is_bipartite(craw2._net)
```

```python
#add shortest path attribute
shortestPath70s = nx.single_source_shortest_path_length(craw2._net, '100')
nx.set_node_attributes(craw2._net, shortestPath70s, 'BipartiteShortestPathTo70sNode')
```

```python
#Add bipartite clustering attribute
bip_clust_coef = nx.algorithms.bipartite.cluster.clustering(craw2._net)
nx.set_node_attributes(craw2._net, bip_clust_coef, 'BipartiteClusteringCoefficient')
```

```python
#Add bipartite closeness centrality. Might take several minutes to calculate
bip_cent_closeness = nx.algorithms.bipartite.centrality.closeness_centrality(craw2._net, top_nodes)
nx.set_node_attributes(craw2._net, bip_cent_closeness, 'BipartiteClosenessCentrality')
```

```python
#Add bipartite degree centrality. Might take several minutes to calculate
bip_cent_degree = nx.algorithms.bipartite.centrality.degree_centrality(craw2._net, top_nodes)
nx.set_node_attributes(craw2._net, bip_cent_degree, 'BipartiteDegreeCentrality')
```

```python
#Add bipartite betweenness centrality. Might take several minutes to calculate
bip_cent_bet = nx.algorithms.bipartite.centrality.betweenness_centrality(craw2._net, top_nodes)
nx.set_node_attributes(craw2._net, bip_cent_bet, 'BipartiteBetweennessCentrality')
```

```python
#Calculate robins alexander clustering
#measures an overall level of clustering in the network
#for more information on the topic I looked into https://toreopsahl.com/tnet/two-mode-networks/clustering/
nx.algorithms.bipartite.cluster.robins_alexander_clustering(G=craw2._net)
```

```python
#Calculates the bipartite density
nx.algorithms.bipartite.density(craw2._net, top_nodes)
```

```python
#Calculates the diameter of the full bipartite graph
nx.algorithms.distance_measures.diameter(craw2._net)
```

```python
#Calculate greedy modularity communities of the entire bipartite graph
#new feature of networkx 2.2
#based on a 2004 paper by Clauset, Newman and Moore http://ece-research.unm.edu/ifis/papers/community-moore.pdf
#Calculates the greedy modularity communities for the full graph
from networkx.algorithms import community
from networkx.algorithms.community import greedy_modularity_communities

full70s_communities = list(greedy_modularity_communities(craw2._net))
len(full70s_communities)
```

```python
#Adds the community attribute to every node
bipartiteCommunityDict = {}
communityNum = 1
for community in full70s_communities:
    for nodename in community:
        bipartiteCommunityDict[nodename] = communityNum
    communityNum = communityNum + 1
nx.set_node_attributes(craw2._net, bipartiteCommunityDict, 'BipartiteCommunity')
```

```python
#Creates two weighted projected graphs
artist70s_nodes = [node for node, node_data in craw2._net.nodes(data=True) if node_data['bipartite']==False]
tags70s_nodes = [node for node, node_data in craw2._net.nodes(data=True) if node_data['bipartite']==True]
artists70s = nx.bipartite.weighted_projected_graph(craw2._net, artist70s_nodes)
tags70s = nx.bipartite.weighted_projected_graph(craw2._net, tags70s_nodes)
```

```python
#Calculates average clustering (approximation)
from networkx.algorithms import approximation

aclust = nx.algorithms.approximation.clustering_coefficient.average_clustering(artists70s)
tclust = nx.algorithms.approximation.clustering_coefficient.average_clustering(tags70s)
print("Average clustering for Artist-to-Artist 70s :")
print(aclust)
print("Average clustering for Tag-to-Tag 70s :")
print(tclust)
```

```python
#Calculate density for both projected graphs
adens = nx.density(artists70s)
tdens = nx.density(tags70s)
print("Artist-to-Artist 70s Density: ")
print(adens)
print("Tag-to-Tag 70s Density: ")
print(tdens)
```

```python
#Calculate diameter of both projected graphs
adiam = nx.algorithms.distance_measures.diameter(artists70s)
tdiam = nx.algorithms.distance_measures.diameter(tags70s)

print("Artist-to-Artist 70s diameter:")
print(adiam)
print("Tag-to-Tag 70s diameter: ")
print(tdiam)
```

```python
artist70s_communities = list(greedy_modularity_communities(artists70s))
tag70s_communities = list(greedy_modularity_communities(tags70s))

print("Number communities in Artist-to-Artist projection: ")
print(len(artist70s_communities))
print("Number of communities in Tag-to-Tag projection: ")
print(len(tag70s_communities))
```

```python
#Adds the community attribute to every node
artCommunityDict = {}
artcommunityNum = 1
for community in artist70s_communities:
    for nodename in community:
        artCommunityDict[nodename] = artcommunityNum
    artcommunityNum = artcommunityNum + 1
nx.set_node_attributes(craw2._net, artCommunityDict, 'ProjectedCommunity')
nx.set_node_attributes(artists70s, artCommunityDict, 'ProjectedCommunity')

tagCommunityDict = {}
tagcommunityNum = 1
for community in tag70s_communities:
    for nodename in community:
        tagCommunityDict[nodename] = tagcommunityNum
    tagcommunityNum = tagcommunityNum + 1
nx.set_node_attributes(craw2._net, tagCommunityDict, 'ProjectedCommunity')
nx.set_node_attributes(tags70s, tagCommunityDict, 'ProjectedCommunity')
```

```python
#Writes the full bipartite graph to gexf and graphml
nx.write_gexf(craw2._net, "70s_5depth_2000.gexf")
nx.write_graphml(craw2._net, "70s_5depth_2000.graphml")
```

```python
#Writes the individual projected graphs to gexf and graphml
nx.write_gexf(artists70s, "artists70s_5d_2000.gexf")
nx.write_graphml(artists70s, "artists70s_5d_2000.graphml")
nx.write_gexf(tags70s, "tags70s_5d_2000.gexf")
nx.write_graphml(tags70s, "tags70s_5d_2000.graphml")
```
